```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path().resolve().parent  
DATASET_DIR = PROJECT_ROOT / "data" / "LSMN2024"

print("Dataset dir exists:", DATASET_DIR.exists())
```

Load Benign

```python
benign_path = DATASET_DIR / "Benign" / "normal_data.csv"

print("Benign file exists:", benign_path.exists())

benign_df = pd.read_csv(benign_path)
benign_df["label"] = "Benign"
```

Load Malicious

```python
dataframes = [benign_df]

malicious_dir = DATASET_DIR / "Malicious"

for attack_folder in malicious_dir.iterdir():
    if attack_folder.is_dir():
        for csv_file in attack_folder.glob("*.csv"):
            df = pd.read_csv(csv_file)
            df["label"] = attack_folder.name
            dataframes.append(df)
```

Merge

```python
full_df = pd.concat(dataframes, ignore_index=True)

print("Final shape:", full_df.shape)
print(full_df["label"].value_counts())
```
